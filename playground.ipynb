```python
import tt
```

```python
import numpy as np
```

```python
tens = np.empty([3, 4, 5, 6, 7, 8], order="F")

# fill
for i0 in range(tens.shape[0]):
    for i1 in range(tens.shape[1]):
        for i2 in range(tens.shape[2]):
            for i3 in range(tens.shape[3]):
                for i4 in range(tens.shape[4]):
                    for i5 in range(tens.shape[5]):
                        tens[i0, i1, i2, i3, i4, i5] = i0 + i1 + i2 + i3 + i4 + i5
```

```python
tens_tt = tt.vector(tens, 1e-12)
```

```python
tt.vector.to_list(tens_tt)
```
