```python
import numpy as np
```

```python
tens = np.empty([3, 4, 5, 6, 7, 8], order="F")

# fill
for i0 in range(tens.shape[0]):
    for i1 in range(tens.shape[1]):
        for i2 in range(tens.shape[2]):
            for i3 in range(tens.shape[3]):
                for i4 in range(tens.shape[4]):
                    for i5 in range(tens.shape[5]):
                        tens[i0, i1, i2, i3, i4, i5] = i0 + i1 + i2 + i3 + i4 + i5
```

```python
tens = np.empty([100, 100, 100], order="F")

# fill
for i0 in range(tens.shape[0]):
    for i1 in range(tens.shape[1]):
        for i2 in range(tens.shape[2]):
            tens[i0, i1, i2] = i0 + 2.71*i1 + 3.14*i2
```

```python
tens = np.empty([3, 4], order="F")

# fill
for i0 in range(tens.shape[0]):
    for i1 in range(tens.shape[1]):
        tens[i0, i1] = i0 + i1
```

```python
tens_tt = tt.vector(tens, 1e-14)
tens_tt
```

```python
l = tt.vector.to_list(tens_tt)
```

```python
for i, x in enumerate(l):
    print(f"core {i}\nshape={x.shape}") #\n{x}")
```

```python
rand_tens = np.random.rand(100, 100, 100)
rand_tt = tt.vector(rand_tens, 1e-14)

for i, x in enumerate(tt.vector.to_list(rand_tt)):
    print(f"core {i}\nshape={x.shape}") #\n{x}")
```
